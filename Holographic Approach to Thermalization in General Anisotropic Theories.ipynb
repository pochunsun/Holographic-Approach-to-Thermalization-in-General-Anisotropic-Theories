* ## The Holography Background

To make our analysis of thermalization as general as possible, we consider the gravitational collapse that eventually forms the following black brane with the metric
$$\tag{1}
ds^2=g_{\mu\nu}dx^{\mu}dx^{\nu}=-y^{-a_0}h(y)d^2t+\frac1{y^{4-a_y}h(y)}d^2y+\sum_{i=1}^{d-1}y^{-a_i}d^2x_i \,,
\label{metric}
$$
where near the boundary $y=0$, the blackening factor is assumed to be $h(0)= 1$. Then, in the boundary, the metric has scaling symmetries,
$$\tag{2}
 y \rightarrow \lambda^{-1} y,~~t \rightarrow \lambda^{1-\frac{a_y}{2}-\frac{a_0}{2}} t,~~x_i \rightarrow \lambda^{1-\frac{a_y}{2}-\frac{a_i}{2}}
 x_i,~~g_{\mu\nu}\rightarrow \lambda^{a_y-2} g_{\mu\nu}\,.
$$
We also assume that there is a simple zero for $h(y)$ at $y=y_h$,
corresponding to the position of the horizon. Thus, near the boundary,
$h(y)$ has the leading term,
$$\tag{3}
  h(y)=1-M y^{{\Delta_h}}
$$
with
$$\tag{4}
{\Delta_h}\geq 1\, .
$$
The temperature of the black brane is given by $T=\frac{|h'(y_h)|}{4\pi
y_h^{\frac{a_0}{2}+\frac{a_y}{2}-2}}$. In order to describe the
formation of the black brane of (\ref{metric}), we introduce the
Eddington-Finkelstein coordinates given by
$$\tag{5}
v=t-\int_0^y\frac{y'^{\frac{a_y}{2}+\frac{a_0}{2}-2}}{h(y')}dy' \, .
$$
Then the metric in (\ref{metric}) becomes
$$\tag{6}
 ds^2=-y^{-a_0}h(y)d^2 v-2
   y^{\frac{a_y}{2}-\frac{a_0}{2}-2}dvdy+\sum_{i=1}^{d-1}y^{a_i}dx_i^2 \, .
$$
In this work, we consider the gravitational collapse with the following Vaidya-type metric
$$\tag{7}
ds^2=-y^{-a_0}f(v,y)d^2 v-2
   y^{\frac{a_y}{2}-\frac{a_0}{2}-2}dvdy+\sum_{i=1}^{d-1}y^{a_i}dx_i^2 \, ,
$$
where $f(v,y)=1-\Theta(v)g(y)$, and $g(y)=1-h(y)$. The metric describes
the formation of the black brane of (\ref{metric}) by  the infalling
of a $d-1$-dimensional delta-function shell along the trajectory
$v=0$. The region $v>0$  outside the shell has the metric as in
(\ref{metric}) whereas the region $v<0$  inside the shell has the pure hyperscaling violating anisotropic Lifshitz metric of the form  (\ref{metric}) by setting $h=1$.
From the viewpoint of the boundary theory,
this gives a quench on the system at $t=0$ and subsequently the system evolves into a thermal equilibrium state.

Here we define $\Delta_0=a_0+a_y-4$ and $\Delta=a_1+a_y-4$ to highlight the nonzero $\Delta_0$ and $\Delta$ effects.

```python
a1=2
a0=2
ay=2
deltap=a0+a1-4
delta=ay+a1-4
d=4
ah=(d-1)*a1/2+deltap/2+1
T=ah/(4*np.pi)
ce=np.sqrt(4*np.pi*T/(a1*(d-1)))
def g(y):
    return y**ah
def h(y):
    return 1-g(y)
```

* ## Dynamics of The Extremal Surface for Taking the Strip as the Boundary

We now consider the entanglement region $\Sigma$ of a strip extending along say $x_1$-direction from $-R$ to $R$ while other $x_i$'s from $-W$ to $W$ with $W\rightarrow\infty$.
In what follows, we denote  $x_1$ as $x$ for simplifying the notation. In this case, the area of the extremal surface $\Gamma$ is expressed as
$$\label{25s}\tag{8}
     A_{\Gamma}= A_{\partial \Sigma}\int_{0}^{R}dx\frac{\sqrt{Q}}{y^{\frac{a_1}{2}(\tilde{d}-1)}}
$$
 with
$$\label{qs}\tag{9}
    Q=\Big\vert 1-\frac{\dot{v}^2 f(y,v)}{y^{\Delta_0-\Delta}}-\frac{2 \dot{v} \dot{y}}{y^{\frac{\Delta_0}{2}-\Delta}}\Big\vert \,.
$$
We define $\dot v=\frac{d}{dx}v(x)$, $\dot
 y=\frac{d}{dx}y(x)$, $A_{\partial \Sigma}=2W^{d-2}$, and $\tilde{d}=\sum_{i=1}^{d-1}\frac{a_i}{a_1}+1$ that can be treated as an effective bulk spacial dimension.

Varying $A_{\Gamma}$ with respect to  the functions $v(x)$ and $y(x)$ leads to  the following equations of motion,
$$\label{eqmot1s}\tag{10}
    \sqrt{Q} y^{\Delta_0-\Delta+\frac{a_1}{2}(\tilde{d}-1)}\frac{d}{dx}\left(\frac{y^{-\frac{a_1}{2}(\tilde{d}-1)-\Delta_0+\Delta}}{\sqrt{Q}} \left(\dot{v} f(y,v)+y^{\frac{\Delta_0}{2}} \dot{y}\right)\right)=\frac{1}{2}\dot{v}^2 \frac{\partial f(y,v)}{\partial v} \, ,
$$
$$\label{eqmot2s}\tag{11}
    \sqrt{Q} y^{\Delta_0-\Delta+\frac{a_1}{2}(\tilde{d}-1)}\frac{d}{dx}\left(\frac{ y^{\frac{1}{2} \left(-a_1 (\tilde{d}-1)-\Delta_0+2\Delta\right)}}{\sqrt{Q}} \dot{v}\right)
    =\frac{1}{2}\left(a_1 (\tilde{d}-1) Q y^{\Delta_0-\Delta-1}-\left(\Delta_0-2\Delta\right) y^{\frac{1}{2} \left(\Delta_0-2\right)} \dot{v} \dot{y}+\frac{\dot{v}^2}{y} \left(\left(-\Delta_0+\Delta\right) f(y,v)+y \frac{\partial f(y,v)}{\partial y}\right)\right) \,.
$$
The translational symmetry in (\ref{25s}) of $A_{\Gamma}$ in the variable $x$ gives the conserved quantity
$$\label{3.20}\tag{12}
 y^{\frac{a_1}{2}(\tilde{d}-1)}\sqrt{Q}= J= {\rm {constant}}=y_t^{\frac{a_1}{2}(\tilde{d}-1)}\, .
$$
The value of $J$ can be determined by the boundary condition at $x=0$, which is the tip of the extremal surface  $y(0)=y_t$.
Again, for $\partial_{v}f=0$ in both $v<0$ and $v>0$ regions, there exists another conserved quantity $E$ given by (\ref{eqmot1s}). Together with (\ref{3.20}), we have
$$\label{ess}\tag{13}
y^{-\Delta_0+\Delta} \left(\dot{v} f(y,v)+y^{\frac{\Delta_0}{2}} \dot{y}\right)=E={\rm {constant}}.
$$
In the vacuum region with $ v<0$, the value of $E$ can be determined at a particular point $x=0$, where the boundary conditions give $E=0$.  Also, $f=1$ in the vacuum region leads to the relation between $\dot{v}$ and $\dot{y}$ at arbitrary $x$ to be
$$\tag{14}
\label{60s}\dot{v}=-y^{\frac{\Delta_0}{2}}\dot{y}\,. 
$$
Substituting all above relations into (\ref{3.20}), for $v<0$ and  $x>0$ and with no loss of generality, it implies
$$\tag{15}
  \label{slope}
  \dot y=-\frac{\sqrt{y_t^{a_1(\tilde{d}-1)}-y^{a_1(\tilde{d}-1)}}}{y^{\frac{1}{2}(a_1(\tilde{d}-1)+\Delta)}}\, .
$$
However, requiring ${dx}/{dy}$ to be finite as $y\rightarrow 0$ gives the constraints
$$\label{uv constraint strip}\tag{16}
      a_1(\tilde{d}-1)>0~~~\text{and}~~~ a_1(\tilde{d}-1)\geq-\Delta \, .
$$
The matching conditions in this case determine the constant $E$ of (\ref{ess}) for the $v>0$ region in terms of the properties of $y(x)$ at $x_c$, given by
$$\label{ebhs}\tag{17}
{E=\frac{g_c}{2}y_c^{\Delta-\Delta_0/2} \dot{y}_-}\,.
$$
Then, via  (\ref{ess}), the relation between $\dot{v}$ and $\dot{y}$ in the black brane region becomes
$$\label{61s}\tag{18}
\dot{v}=\frac{1}{h}\left(-\dot{y} y^{\frac{\Delta_0}{2}}+E y^{\Delta_0-\Delta}\right)\, .
$$
The equation for $y(x)$
in the black brane region with $v>0$ as in (\ref{slope}) can be found from (\ref{61s}) with the relation (\ref{3.20}) as
$$\label{3.36s}\tag{19}
 \dot{y}^{2}=H(y)
 =h(y) \Big(\frac{y_{t}^{a_1(\tilde{d}-1)}}{y^{a_1(\tilde{d}-1)}}-1\Big)y^{-\Delta}+E^{2}y^{\Delta_0-2\Delta}\, .
$$
 With (\ref{61s}) and the square root of (\ref{3.36s}), we obtain for $x>0$
$$\label{3.37s}\tag{20}
        \frac{dv}{dy}= \frac{\dot{v}}{-\sqrt{H(y)}}
        =-\frac{1}{h(y)}\Big(y^{\frac{\Delta_0}{2}}+\frac{Ey^{\Delta_0-\Delta}}{\sqrt{H(y)}}\Big)\, .
$$
 In the end, from (\ref{slope}) and (\ref{3.36s}), we can write down the relation between the strip width $R$ and the values of $y_c$ and $y_t$ as
$$\label{strip R}\tag{21}
 R=\int_{y_c}^{y_t}\frac{y^{\frac{a_1}{2}(\tilde{d}-1)+\frac{\Delta}{2}}}{y_t^{\frac{a_1}{2}(\tilde{d}-1)}\sqrt{1-\frac{y^{a_1(\tilde{d}-1)}}{y_t^{a_1(\tilde{d}-1)}}}}\,dy+\int_{0}^{y_c}\frac{1}{\sqrt{H(y)}}\,dy \,.
$$
  From (\ref{3.37s}) and the boundary condition $v(R)=t$, $y_c$ can also be expressed as an implicit function of the boundary time $t$,
$$\label{time_strip_text}\tag{22}
  t=\int_{0}^{y_c}\frac{1}{h(y)}\Big(y^{\frac{\Delta_0}{2}}+\frac{Ey^{\Delta_0-\Delta}}{\sqrt{H(y)}}\Big)\,dy \, .
$$
Moreover,
substituting (\ref{slope}), (\ref{61s}), (\ref{60s}) and (\ref{3.36s}) into (\ref{qs}), we then write the integral formula
 for the area in (\ref{25s}) in terms of the contributions from the $v<0$ and $v>0$ regions respectively as
$$\label{area_strip}\tag{23}
\frac{A_\Gamma}{A_{\partial\Sigma}}= \frac{A_{\Gamma v<0}}{A_{\partial\Sigma}}+\frac{A_{\Gamma v>0}}{A_{\partial\Sigma}}=\int_{y_c}^{y_t}\frac{1}{y^{\frac{a_1}{2}(\tilde{d}-1)-\frac{\Delta}{2}}\sqrt{1-\frac{y^{a_1(\tilde{d}-1)}}{y_t^{a_1(\tilde{d}-1)}}}}\,dy+\int_{0}^{y_c}\frac{y_t^{\frac{a_1}{2}(\tilde{d}-1)}}{y^{a_1(\tilde{d}-1)}\sqrt{H(y)}}\,dy
\, .$$
Then, through  (\ref{strip R}) and (\ref{time_strip_text}), the area are expressed in terms of the boundary time $t$ and the strip width $R$.
Notice that with the additional constant $J$ in the strip case,  the area in (\ref{25s}) and the corresponding entanglement entropy can be  computed  just by the values of $y(x)$ at $x=0$ and $x=x_c$.

* ## Important Results for the Strip

### **<font color = orange>1. Entanglement Growth in Early Times</font>**

In terms of the boundary time, the early time entanglement entropy growth for both small and large $R$ limits is given by
$$\label{early growth}\tag{A.1}
\delta S(t)\simeq \frac{\delta A_{\Gamma}(t)}{4}\simeq \frac{A_{\partial
\Sigma}M}{4(\Delta+2{\Delta_h}+2-a_1(\tilde{d}-1))}\left(\frac{\Delta_0}{2}+1\right)^{\frac{\Delta+2{\Delta_h}+2-a_1(\tilde{d}-1)}{\Delta_0+2}}t^{\frac{\Delta+2{\Delta_h}+2-a_1(\tilde{d}-1)}{\Delta_0+2}}
\,$$
where
$$\label{S_c_sphere}
\Delta+2{\Delta_h}+2-a_1(\tilde{d}-1) >0
$$
to ensure that the entanglement entropy increases in time. In particular, for the positive (negative) value of $\Delta$, the power of $t$ increases (decreases) with $\Delta$  ($|\Delta|$) that speeds up (slows down) the growth rates as compared with the case of $\Delta=0$.

### **<font color = orange>2. Saturation Time in the Small $R $ Limit</font>**

In the small $R $ limit, the relation between the tip of the extremal surface $y_b$ and the size of the boundary $R$ in (\ref{strip R}) allow us to write down the approximate saturation time $t_s$
$$\label{ts strip smallr_test}\tag{A.2}
t_s\simeq \frac{2}{\Delta_0+2}\left(\frac{a_1(\tilde{d}-1)}{\tilde{\mathscr{R}}_0}\right)^{\frac{\Delta_0+2}{\Delta+2}}R^{\frac{\Delta_0+2}{\Delta+2}} \,
$$
with 
$$
\tilde{\mathscr{R}}_0=\frac{\Gamma \left(\frac{1}{2}\right) \Gamma\left(\frac{1}{2}+\frac{\Delta+2}{2a_1(\tilde{d}-1)}\right)}{\Gamma\left(1+\frac{\Delta+2}{2a_1(\tilde{d}-1)}\right)}.
$$
Again, the contribution of  $\Delta \neq 0$ to the powers of $R$ can be checked from the saturation time $t_s$ obtained in the field theories.
For a positive (negative) $\Delta$, in the small $R$ limit, the saturation time
$t_s$ becomes larger (smaller) than the case of $\Delta=0$  for  fixed $R$ and $T$.

### **<font color = orange>3. Saturation Time in the Large $R $ Limit</font>**

$$\label{ts}\tag{A.3}
t_s=\int_0^{y_b}dy \frac{y^{\frac{\Delta_0}{2}}}{h(y)}=\frac{R}{c_E}+O(R^0)~~~\text{with $c_E=\sqrt{\frac{4\pi T y_h^{\Delta+1-\Delta_0/2}}{a_1(d-1)}}$} \, .$$
Here $T$ is the black brane temperature defined above, and the  above integral is dominated in the near horizon region.
In terms of $T$, $c_E \propto T^{\frac{\Delta_0-\Delta}{\Delta_0+2}}$. The saturation time $t_s \propto R$ in the large $R$ limit. For a positive (negative) $\Delta$, $c_E$ becomes smaller (larger) than the case of $\Delta=0$, leading to relatively larger (smaller) the saturation time scale $t_s$ for fixed $R$ and $T$.

```python
#def of yb(R)
def yb(target_R):
    errorR=100
    accep_error=10**(-6)
    yb_lower=0
    yb_upper=1
    while abs(errorR) > accep_error:
        ybb=(yb_lower+yb_upper)/2
        test_R=integrate.quad(lambda y: (y**((delta+a1*(d-1))/2))/((ybb**(a1*(d-1)/2))*np.sqrt(h(y))*np.sqrt(1-(y/ybb)**(a1*(d-1)))), 0, ybb)
        if test_R[0] > target_R :
            yb_upper=ybb
        else: 
            yb_lower=ybb
        errorR=(target_R-test_R[0])/target_R
    return (ybb,test_R[1])
#def of ts
def ts(ybb):
    return integrate.quad(lambda y: (y**(deltap/2))/h(y), 0, ybb)

#def of ts for small R
def ts_small_pre(R):
    return (2/(deltap+2))*((a1*(d-1)*R/special.beta((delta+2)/(2*a1*(d-1))+1/2, 1/2))**((deltap+2)/(delta+2)))

#def of aeq
def aeq(ybb,yt00):
    Avac0=integrate.quad(lambda y: 1/((y**((a1*(d-1)-delta)/2))*np.sqrt(1-(y/yt00)**(a1*(d-1)))),auv, yt00)
    Abh=integrate.quad(lambda y:  1/((y**((a1*(d-1)-delta)/2))*np.sqrt(h(y))*np.sqrt(1-(y/ybb)**(a1*(d-1)))),auv, ybb)
    return (Abh[0]-Avac0[0],Abh[1]+Avac0[1])

b=ts(yb(2)[0])[0]-2/ce
```

```python
R_range = np.arange(0.0, 2.01, 0.01) #R from 0 to 2
ts_list=[0]
for i in range(1, 201): #i from 0 to 200
    ts_list.append(ts(yb(R_range[i])[0])[0])

#large R prediction
largeR_range = np.arange(0.0, 2.01, 0.01) #R from 0 to 2
largeR_ts_list=[b]
for i in range(1, 201): #i from 0 to 200
    largeR_ts_list.append(b+largeR_range[i]/ce)

#small R prediction
smallR_range = np.arange(0.0, 2.01, 0.01) #R from 0 to 2
smallR_ts_list=[0]
for i in range(1, 201): #i from 0 to 200
    smallR_ts_list.append(ts_small_pre(smallR_range[i]))
```

* ## Dynamics of The Extremal Surface (Numerical)

```python
#conserved quantity E
def E(ycc,ytt):
    return -g(ycc)*(ycc**(delta-deltap/2))*np.sqrt(ytt**(a1*(d-1))-ycc**(a1*(d-1)))/(2*ycc**((a1*(d-1)+delta)/2))

#yt0
def yt0(target_R):
    errorR=100
    accep_error=10**(-6)
    yt_lower=0
    yt_upper=5*target_R
    while abs(errorR) > accep_error:
        ytt=(yt_lower+yt_upper)/2
        test_R=integrate.quad(lambda y: (y**((delta+a1*(d-1))/2))/((ytt**(a1*(d-1)/2))*np.sqrt(1-(y/ytt)**(a1*(d-1)))), 0, ytt)
        if test_R[0] > target_R :
            yt_upper=ytt
        else: 
            yt_lower=ytt
        errorR=(target_R-test_R[0])/target_R
    return (ytt,test_R[1])

def He(y,ycc,ytt):
    return h(y)*((ytt/y)**(a1*(d-1))-1)*(y**(-delta))+(E(ycc,ytt)**2)*(y**(deltap-2*delta))

#def of time
def t(ycc,ytt):
    return integrate.quad(lambda y: (y**(deltap/2)+E(ycc,ytt)*(y**(deltap-delta))/np.sqrt(He(y,ycc,ytt)))/h(y), 0, ycc)

#def of R #function of yc yt
def yt(target_R,ycc):
    errorR=100
    accep_error=10**(-6)
    yt_lower=ycc
    yt_upper=yt0(target_R)[0]
    while abs(errorR) > accep_error:
        ytt=(yt_lower+yt_upper)/2
        Rvac=integrate.quad(lambda y: (y**((delta+a1*(d-1))/2))/((ytt**(a1*(d-1)/2))*np.sqrt(1-(y/ytt)**(a1*(d-1)))), ycc, ytt)
        Rbh=integrate.quad(lambda y: 1/np.sqrt(He(y,ycc,ytt)), 0, ycc)
        test_R=(Rvac[0]+Rbh[0],Rvac[1]+Rbh[1])
        if test_R[0] > target_R :
            yt_upper=ytt
        else: 
            yt_lower=ytt
        errorR=(target_R-test_R[0])/target_R
    return (ytt,test_R[1])


#def of area #function of yc yt
def area(ycc,ytt,yt00):
    Avac=integrate.quad(lambda y: 1/((y**((a1*(d-1)-delta)/2))*np.sqrt(1-(y/ytt)**(a1*(d-1)))),ycc, ytt)
    Avac0=integrate.quad(lambda y: 1/((y**((a1*(d-1)-delta)/2))*np.sqrt(1-(y/yt00)**(a1*(d-1)))),auv, yt00)
    Abh=integrate.quad(lambda y: (ytt**(a1*(d-1)/2))/((y**(a1*(d-1)))*np.sqrt(He(y,ycc,ytt))), auv, ycc)
    return (Avac[0]+Abh[0]-Avac0[0],Avac[1]+Abh[1]+Avac0[1])

#def of late time
def area_late_time(tt,tss,aeqq,ybb):
    I=integrate.quad(lambda y: (y**(deltap-delta/2))/((h(y)**(3/2))*np.sqrt(-1+(ybb/y)**(a1*(d-1)))),0, ybb)
    return (4*(tt-tss)/(a1*(d-1)*(1+h(ybb))*I[0]))*(ybb**(deltap/2+1-a1*(d-1)/2))+aeqq

#def of early time
def area_early_time(tt):
    return (((deltap/2+1)*tt)**((deltap+delta+4)/(deltap+2)))/(deltap+delta+4)

#ycdotp
def ycdotp(ycc,ytt):
    return -(1-g(ycc)/2)*((ytt**(a1*(d-1))-ycc**(a1*(d-1)))**(1/2))/(ycc**((a1*(d-1)+delta)/2))
#ym
def ym(ycc,ytt):
    y = ycc  # Starting point
    step_size = 0.0001
    step_tolerance = 10**(-7)
    max_iterations = 10**(6)
    # Derivative function H'(y)
    dH=lambda y: (-2*delta+deltap)*(E(ycc,ytt)**2)*(y**(-1-2*delta+deltap))-a1*(-1+d)*(y**(-2-delta))*(1-y**ah)*ytt*((ytt/y)**(-1+a1*(-1+d)))-ah*(y**(-1+ah-delta))*(-1+(ytt/y)**(a1*(-1+d)))-delta*(y**(-1-delta))*(1-y**ah)*(-1+(ytt/y)**(a1*(-1+d)))
    # Gradient descent iteration
    for j in range(max_iterations):
        step = step_size * dH(y)
        y -= step
        if y >= ytt:
            y=0
            break
        if abs(step) <= step_tolerance:
            break
    return y-step_tolerance
#yr
def yr(ymm,ycc,ytt):
    erroryr=100
    accep_error=10**(-7)
    yr_lower=ycc
    yr_upper=ymm
    while abs(erroryr) > accep_error or He(yrr,ycc,ytt) < 0:
        yrr=(yr_lower+yr_upper)/2
        if He(yrr,ycc,ytt) < 0 :
            yr_upper=yrr
        else: 
            yr_lower=yrr
        erroryr=He(yrr,ycc,ytt)
    return (yrr,erroryr)
```

```python
#def of modified R #function of yc yt
def ytm(target_R,ycc,yrr):
    errorR=100
    accep_error=10**(-6)
    yt_lower=ycc
    yt_upper=yt0(target_R)[0]
    while abs(errorR) > accep_error:
        ytt=(yt_lower+yt_upper)/2
        Rvac=integrate.quad(lambda y: (y**((delta+a1*(d-1))/2))/((ytt**(a1*(d-1)/2))*np.sqrt(1-(y/ytt)**(a1*(d-1)))), ycc, ytt)
        Rbh1=integrate.quad(lambda y: 1/np.sqrt(He(y,ycc,ytt)), 0, ycc)
        Rbh2=integrate.quad(lambda y: 2/np.sqrt(He(y,ycc,ytt)), ycc,yrr-abs(yrr-ycc)*10**(-2))
        test_R=(Rvac[0]+Rbh1[0]+Rbh2[0],Rvac[1]+Rbh1[1]+Rbh2[1])
        if test_R[0] > target_R :
            yt_upper=ytt
        else: 
            yt_lower=ytt
        errorR=(target_R-test_R[0])/target_R
    return (ytt,test_R[1])

#def of modified time
def tm(ycc,ytt,yrr):
    tm1=integrate.quad(lambda y: (y**(deltap/2)+E(ycc,ytt)*(y**(deltap-delta))/np.sqrt(He(y,ycc,ytt)))/h(y), 0, ycc)
    tm2=integrate.quad(lambda y: 2*(y**(deltap/2)+E(ycc,ytt)*(y**(deltap-delta))/np.sqrt(He(y,ycc,ytt)))/h(y),ycc,yrr-abs(yrr-ycc)*10**(-2))
    return (tm1[0]+tm2[0],tm1[1]+tm2[1])

#def of modified area #function of yc yt
def aream(ycc,ytt,yt00,yrr):
    Avac=integrate.quad(lambda y: 1/((y**((a1*(d-1)-delta)/2))*np.sqrt(1-(y/ytt)**(a1*(d-1)))),ycc, ytt)
    Avac0=integrate.quad(lambda y: 1/((y**((a1*(d-1)-delta)/2))*np.sqrt(1-(y/yt00)**(a1*(d-1)))),auv, yt00)
    Abh1=integrate.quad(lambda y: (ytt**(a1*(d-1)/2))/((y**(a1*(d-1)))*np.sqrt(He(y,ycc,ytt))), auv, ycc)
    Abh2=integrate.quad(lambda y: 2*(ytt**(a1*(d-1)/2))/((y**(a1*(d-1)))*np.sqrt(He(y,ycc,ytt))), ycc, yrr-abs(yrr-ycc)*10**(-2))
    return (Avac[0]+Abh1[0]+Abh2[0]-Avac0[0],Avac[1]+Abh1[1]+Abh2[1]+Avac0[1])
```

```python
#small R
auv=0.0005
R_test=0.1
time=0
ycc=0
yt00=yt0(R_test)[0]
ytt=yt00
yt_list=[ytt]
yc_list=[0]
time_list=[0]
area_list=[0]
i=100
for j in range(5,232): #ts(yb) and #yb(R)
    ycc=j*0.001
    ytt=yt(R_test,ycc)[0]
    if ycc+0.001 >= yt00 :
        break
    if ycc >= ytt :
        break
    if ycdotp(ycc+10**(-4),ytt) >= 0 :
        break
    At=area(ycc,ytt,yt00)[0] #A(yc,yt)
    time=t(ycc,ytt)[0] #t(yc,yt)
    if ts(ytt)[0]<time : #ts(ybb)
        break
    area_list.append(At)
    time_list.append(time)
    yc_list.append(ycc)
    yt_list.append(ytt)
    
#
tsexper=time
final_E=E(ycc,ytt)
ybb=yb(R_test)[0]
tss=ts(ybb)[0]
#aeqq=aeq(ybb,yt00)[0]
for j in range(0,20): #ts(yb) and #yb(R)
    time=tss+j*0.01
     #At=aeqq #A(yc,yt)
    time=time#t(yc,yt)
    area_list.append(At)
    time_list.append(time)
```
